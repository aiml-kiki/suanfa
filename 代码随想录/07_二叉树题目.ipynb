 1

[4,7,2,9,6,3,1]

 2
root = [2,1,3]
[2,3,1]

 3
root = []
[]
```

````python
class Solution:
    def invertTree(self, root):

        def ress(root):
            if not root:
                return
            tmp = root.left
            root.left = root.right
            root.right = tmp
            ress(root.left)
            ress(root.right)
            return root
        
        return ress(root)
````

# [](https://leetcode-cn.com/problems/maximum-depth-of-binary-tree/)





****: 
```

 [3,9,20,null,null,15,7]
    3
   / \
  9  20
    /  \
   15   7
 3 
```

````python
class Solution:
    def maxDepth(self, root):
        if not root:
            return 0
        
        res = 0
        dq = deque([root])
        while dq:
            n = len(dq)
            for _ in range(n):
                node = dq.popleft()
                if node.left:
                    dq.append(node.left)
                if node.right:
                    dq.append(node.right)
            res += 1
        return res
````

#  [](https://leetcode-cn.com/problems/count-complete-tree-nodes/)

```root`

  h `1~ 2h`

```
 1
root = [1,2,3,4,5,6]
6

 2
root = []
0

 3
root = [1]
1
```

````python
class Solution:
    def countNodes(self, root):
        
        if not root:
            return 0
        res = 1
        dq = deque([root])
        while dq:
            node = dq.popleft()
            if node.left:
                dq.append(node.left)
                res += 1
            if node.right:
                dq.append(node.right)
                res += 1
        return res
````

# [](https://leetcode-cn.com/problems/balanced-binary-tree/)




 `1`




```
 1
root = [3,9,20,null,null,15,7]
true

 2
root = [1,2,2,3,3,null,null,4,4]
false

 3
root = []
true
```

````python
class Solution:
    def isBalanced(self, root):
        return self.recur(root) != -1

    def recur(self, root):
        if not root: 
            return 0
        
        left = self.recur(root.left)
        if left == -1:
             return -1
        right = self.recur(root.right)
        if right == -1: 
            return -1
        return max(left, right) + 1 if abs(left - right) < 2 else -1
````

# [](https://leetcode-cn.com/problems/binary-tree-paths/)

`root```

 

`` 
 1
root = [1,2,3,null,5]
["1->2->5","1->3"]

 2
root = [1]
["1"]
```

````python
class Solution:
    def binaryTreePaths(self, root):
        
        if not root:
            return []
        
        res = []
        dq = deque([root])
        pat = deque([str(root.val)])
        while dq:
            node = dq.pop()
            path = pat.pop()
            if node.left:
                pat.append(path + '->' + str(node.left.val))
                dq.append(node.left)
            if node.right:
                pat.append(path + '->' + str(node.right.val))
                dq.append(node.right)
            if not node.left and not node.right:
                res.append(path)
        
        return res
````
