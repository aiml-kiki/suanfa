#  [翻转二叉树](https://leetcode-cn.com/problems/invert-binary-tree/)

给你一棵二叉树的根节点`root`，翻转这棵二叉树，并返回其根节点。

```
示例 1：
输入：root = [4,2,7,1,3,6,9]
输出：[4,7,2,9,6,3,1]

示例 2：
输入：root = [2,1,3]
输出：[2,3,1]

示例 3：
输入：root = []
输出：[]
```

````python
class Solution:
    def invertTree(self, root):

        def ress(root):
            if not root:
                return
            tmp = root.left
            root.left = root.right
            root.right = tmp
            ress(root.left)
            ress(root.right)
            return root
        
        return ress(root)
````

# [二叉树的最大深度](https://leetcode-cn.com/problems/maximum-depth-of-binary-tree/)

给定一个二叉树，找出其最大深度。

二叉树的深度为根节点到最远叶子节点的最长路径上的节点数。

**说明**: 叶子节点是指没有子节点的节点。
```
示例：
给定二叉树 [3,9,20,null,null,15,7]，
    3
   / \
  9  20
    /  \
   15   7
返回它的最大深度 3 。
```

````python
class Solution:
    def maxDepth(self, root):
        if not root:
            return 0
        
        res = 0
        dq = deque([root])
        while dq:
            n = len(dq)
            for _ in range(n):
                node = dq.popleft()
                if node.left:
                    dq.append(node.left)
                if node.right:
                    dq.append(node.right)
            res += 1
        return res
````

#  [完全二叉树的节点个数](https://leetcode-cn.com/problems/count-complete-tree-nodes/)

给你一棵`完全二叉树`的根节点`root`，求出该树的节点个数。

完全二叉树 的定义如下：在完全二叉树中，除了最底层节点可能没填满外，其余每层节点数都达到最大值，并且最下面一层的节点都集中在该层最左边的若干位置。若最底层为第 h 层，则该层包含`1~ 2h`个节点。

```
示例 1：
输入：root = [1,2,3,4,5,6]
输出：6

示例 2：
输入：root = []
输出：0

示例 3：
输入：root = [1]
输出：1
```

````python
class Solution:
    def countNodes(self, root):
        
        if not root:
            return 0
        res = 1
        dq = deque([root])
        while dq:
            node = dq.popleft()
            if node.left:
                dq.append(node.left)
                res += 1
            if node.right:
                dq.append(node.right)
                res += 1
        return res
````

# [平衡二叉树](https://leetcode-cn.com/problems/balanced-binary-tree/)


给定一个二叉树，判断它是否是高度平衡的二叉树。
本题中，一棵高度平衡二叉树定义为：
一个二叉树每个节点 的左右两个子树的高度差的绝对值不超过`1`。




```
示例 1：
输入：root = [3,9,20,null,null,15,7]
输出：true

示例 2：
输入：root = [1,2,2,3,3,null,null,4,4]
输出：false

示例 3：
输入：root = []
输出：true
```

````python
class Solution:
    def isBalanced(self, root):
        return self.recur(root) != -1

    def recur(self, root):
        if not root: 
            return 0
        
        left = self.recur(root.left)
        if left == -1:
             return -1
        right = self.recur(root.right)
        if right == -1: 
            return -1
        return max(left, right) + 1 if abs(left - right) < 2 else -1
````

# [二叉树的所有路径](https://leetcode-cn.com/problems/binary-tree-paths/)

给你一个二叉树的根节点`root`，按`任意顺序`，返回所有从根节点到叶子节点的路径。

叶子节点 是指没有子节点的节点。

`` 
示例 1：
输入：root = [1,2,3,null,5]
输出：["1->2->5","1->3"]

示例 2：
输入：root = [1]
输出：["1"]
```

````python
class Solution:
    def binaryTreePaths(self, root):
        
        if not root:
            return []
        
        res = []
        dq = deque([root])
        pat = deque([str(root.val)])
        while dq:
            node = dq.pop()
            path = pat.pop()
            if node.left:
                pat.append(path + '->' + str(node.left.val))
                dq.append(node.left)
            if node.right:
                pat.append(path + '->' + str(node.right.val))
                dq.append(node.right)
            if not node.left and not node.right:
                res.append(path)
        
        return res
````
