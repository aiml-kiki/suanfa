## [](https://leetcode-cn.com/problems/remove-linked-list-elements/)

`head``val``Node.val == val```



```
 1
head = [1,2,6,3,4,5,6], val = 6
[1,2,3,4,5]

 2
head = [], val = 1
[]

 3
head = [7,7,7,7], val = 7
[]
```

````python
class Solution:
    def removeElements(self, head, val):
        cur = ListNode(next=head)
        res = cur
        while cur.next:
            if cur.next.val == val:
                cur.next = cur.next.next
            else:
                cur = cur.next
        return res.next
````

## [](https://leetcode-cn.com/problems/design-linked-list/)

`val``next``val``next`/`prev``0-index`



- get(index)`index`-1
- addAtHead(val)`val`
- addAtTail(val)`val`
- addAtIndex(index,val)`index``val``index``index`index0
- deleteAtIndex(index)`index``index`

````python
class ListNode:

    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next


class MyLinkedList:

    def __init__(self):
        self.head = ListNode()
        self.len = 0

    def get(self, index: int) -> int:
        if 0 <= index < self.len:
            node = self.head
            for _ in range(index + 1):
                node = node.next
            return node.val
        else:
            return -1


    def addAtHead(self, val: int) -> None:
        self.addAtIndex(0, val)


    def addAtTail(self, val: int) -> None:
        self.addAtIndex(self.len, val)


    def addAtIndex(self, index: int, val: int) -> None:
        if index < 0:
            index = 0
        elif index > self.len:
            return

        self.len += 1
        
        add_node = ListNode(val=val)
        pre = None
        cur = self.head
        
        for _ in range(index + 1):
            pre = cur
            cur = cur.next
        pre.next = add_node
        add_node.next = cur

    def deleteAtIndex(self, index: int) -> None:
        if 0 <= index < self.len:
            self.len -= 1
            pre = ListNode()
            cur = self.head
            for index in range(index + 1):
                pre = cur
                cur = cur.next
            pre.next = cur.next
            cur.next = None
````

## [](https://leetcode-cn.com/problems/reverse-linked-list/)

`head`

```
 1
head = [1,2,3,4,5]
[5,4,3,2,1]

 2
head = [1,2]
[2,1]

 3
head = []
[]
```

````python
class Solution:
    def reverseList(self, head):
        pre = None
        cur = head
        while cur:
            tmp = cur.next
            cur.next = pre
            pre = cur
            cur = tmp
        
        return pre
````

## [](https://leetcode-cn.com/problems/swap-nodes-in-pairs/)





```
 1
head = [1,2,3,4]
[2,1,4,3]

 2
head = []
[]

 3
head = [1]
[1]
```

````python
class Solution:
    def swapPairs(self, head):
        tmp = ListNode(next=head)
        pre = tmp
        
        while pre.next and pre.next.next:
            l = pre.next.next
            cur = pre.next

            cur.next = l.next
            l.next = cur
            pre.next = l


            pre = pre.next.next
        
        return tmp.next
````

## [N](https://leetcode-cn.com/problems/remove-nth-node-from-end-of-list/)

`n`

```
 1
head = [1,2,3,4,5], n = 2
[1,2,3,5]

 2
head = [1], n = 1
[]

 3
head = [1,2], n = 1
[1]
```

````python
class Solution:
    def removeNthFromEnd(self, head, n):
        
        tmp =ListNode(next=head)
        fast = tmp
        slow = tmp
        
        for _ in range(n+1):
            fast = fast.next
        
        while fast:
            fast = fast.next
            slow = slow.next
        
        slow.next = slow.next.next

        return tmp.next
````

````python
class Solution:
    def getIntersectionNode(self, headA, headB):
        A = headA
        B = headB
        while A != B:
            A = A.next if A else headB
            B = B.next if B else headA
        return A
````

## [](https://leetcode-cn.com/problems/intersection-of-two-linked-lists-lcci/)

````python
class Solution:
    def detectCycle(self, head):
        slow, fast = head, head
        while fast and fast.next:
            slow = slow.next
            fast = fast.next.next
            if slow == fast:
                p = head
                while p != slow:
                    p = p.next
                    slow = slow.next
                
                return slow
        return
````
